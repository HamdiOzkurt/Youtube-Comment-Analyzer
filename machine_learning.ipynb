```python
import pandas as pd
import time
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score  #Buradaki kod bloƒüunda modelin a≈üƒ±rƒ± √∂ƒürenmeye gitmemesi i√ßin GridSearchCV kullanƒ±ldƒ±.
import warnings
warnings.filterwarnings('ignore') 

def get_models_with_gridsearch(X, y, cv=5, scoring='accuracy', info=True):
    """
    GridSearchCV ile hyperparameter tuning yaparak modelleri eƒüitir.
    Cross-validation ile overfitting riskini minimize eder.
    
    Parameters:
    -----------
    X : feature matrisi
    y : hedef deƒüi≈üken
    cv : cross-validation fold sayƒ±sƒ± (default=5)
    scoring : metrik (default='accuracy')
    info : detaylƒ± bilgi g√∂ster (default=True)
    """
    
    if info:
        print("Veri ayrƒ±lƒ±yor...")
        print(f"X shape: {X.shape}, y shape: {y.shape}")
        print(f"Cross-Validation Fold: {cv}\n")
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
  
    param_grids = {
        'LR': {
            'model': LogisticRegression(max_iter=1000, random_state=42),
            'params': {
                'C': [0.01, 0.1, 1, 10],
                'penalty': ['l1', 'l2'],
                'solver': ['liblinear', 'saga']
            }
        },
        
        'CART': {
            'model': DecisionTreeClassifier(random_state=42),
            'params': {
                'max_depth': [5, 10, 15, 20, None],
                'min_samples_split': [10, 20, 50],
                'min_samples_leaf': [5, 10, 20],
                'max_features': ['sqrt', 'log2', None]
            }
        },
        
        'RF': {
            'model': RandomForestClassifier(random_state=42),
            'params': {
                'n_estimators': [50, 100, 200],
                'max_depth': [10, 15, 20, None],
                'min_samples_split': [10, 20],
                'min_samples_leaf': [5, 10],
                'max_features': ['sqrt', 'log2']
            }
        },
        
        'SVM': {
            'model': SVC(random_state=42),
            'params': {
                'C': [0.1, 1, 10],
                'kernel': ['rbf', 'linear'],
                'gamma': ['scale', 'auto', 0.01, 0.1]
            }
        },
        
        'KNN': {
            'model': KNeighborsClassifier(),
            'params': {
                'n_neighbors': [3, 5, 7, 9, 11],
                'weights': ['uniform', 'distance'],
                'metric': ['euclidean', 'manhattan', 'minkowski']
            }
        },
        
        'GradientBoosting': {
            'model': GradientBoostingClassifier(random_state=42),
            'params': {
                'n_estimators': [50, 100, 150],
                'learning_rate': [0.01, 0.1, 0.2],
                'max_depth': [3, 5, 7],
                'subsample': [0.8, 1.0],
                'min_samples_split': [10, 20]
            }
        },
        
        'XGBoost': {
            'model': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
            'params': {
                'n_estimators': [50, 100, 150],
                'learning_rate': [0.01, 0.1, 0.2],
                'max_depth': [3, 5, 7],
                'subsample': [0.8, 1.0],
                'colsample_bytree': [0.8, 1.0],
                'reg_alpha': [0, 0.1, 1],
                'reg_lambda': [1, 2, 5]
            }
        },
        
        'LightGBM': {
            'model': LGBMClassifier(verbose=-1, random_state=42),
            'params': {
                'n_estimators': [50, 100, 150],
                'learning_rate': [0.01, 0.1, 0.2],
                'max_depth': [3, 5, 7],
                'num_leaves': [15, 31, 63],
                'subsample': [0.8, 1.0],
                'colsample_bytree': [0.8, 1.0],
                'reg_alpha': [0, 0.1, 1],
                'reg_lambda': [1, 2, 5]
            }
        },
        
        'CatBoost': {
            'model': CatBoostClassifier(verbose=False, random_state=42),
            'params': {
                'iterations': [50, 100, 150],
                'learning_rate': [0.01, 0.1, 0.2],
                'depth': [3, 5, 7],
                'l2_leaf_reg': [1, 3, 5],
                'subsample': [0.8, 1.0]
            }
        }
    }
    
    results = []
    best_models = {}
    
    print(f"Toplam {len(param_grids)} model i√ßin GridSearchCV ba≈ülƒ±yor...")
    print(f"{'='*70}\n")
    
    for name, config in param_grids.items():
        if info:
            print(f"üîç {name} i√ßin GridSearch yapƒ±lƒ±yor...")
        
        start_time = time.time()
        
        try:
           
            grid_search = GridSearchCV(
                estimator=config['model'],
                param_grid=config['params'],
                cv=cv,
                scoring=scoring,
                n_jobs=-1,
                verbose=0
            )
            
       
            grid_search.fit(X_train, y_train)
            
          
            best_model = grid_search.best_estimator_
            best_models[name] = best_model
            
         
            y_pred_test = best_model.predict(X_test)
            y_pred_train = best_model.predict(X_train)
            
            
            train_time = time.time() - start_time
            
            
            acc_test = accuracy_score(y_test, y_pred_test)
            precision_test = precision_score(y_test, y_pred_test, average='weighted', zero_division=0)
            recall_test = recall_score(y_test, y_pred_test, average='weighted', zero_division=0)
            f1_test = f1_score(y_test, y_pred_test, average='weighted', zero_division=0)
            
            acc_train = accuracy_score(y_train, y_pred_train)
            
            
            cv_score = grid_search.best_score_
            
          
            overfit_diff = acc_train - acc_test
            
            results.append([
                name, 
                acc_test, 
                acc_train, 
                cv_score,
                overfit_diff, 
                precision_test, 
                recall_test, 
                f1_test, 
                train_time
            ])
            
            if info:
                print(f"Test: %{acc_test*100:.2f} | Train: %{acc_train*100:.2f} | CV: %{cv_score*100:.2f}")
                print(f"En iyi parametreler: {grid_search.best_params_}")
                print(f"S√ºre: {train_time:.2f}s")
                
               
                if overfit_diff > 0.10:
                    print(f"UYARI: Overfitting riski y√ºksek! (Fark: %{overfit_diff*100:.2f})")
                print()
            
        except Exception as e:
            print(f"{name} HATA: {str(e)}\n")
            continue
    
    
    df_results = pd.DataFrame(
        results, 
        columns=['name', 'acc_test', 'acc_train', 'cv_score', 'overfit_diff', 
                'precision', 'recall', 'f1', 'train_time']
    )
    df_results = df_results.sort_values(by='acc_test', ascending=False).reset_index(drop=True)
    
    if info:
        print("\n" + "="*70)
        print("SONU√áLAR (Test Accuracy'ye g√∂re sƒ±ralƒ±):")
        print("="*70)
        print(df_results.to_string(index=False))
        print("\n" + "="*70)
        print("ƒ∞PUCU:")
        print(" - CV Score: Cross-validation skoru (genelleme yeteneƒüi)")
        print(" - Overfit Diff: Train-Test farkƒ± (>%10 ise overfitting riski)")
        print(" - En iyi model: CV Score ve Test Acc dengesi y√ºksek olan")
        print("="*70)
    
    return df_results, best_models
```

```python
import pandas as pd
from timeit import default_timer as timer
from datetime import timedelta
from tqdm import tqdm

def preprocessing(series, remove_hashtag=False, remove_mentions=False, remove_links=False, 
                  remove_numbers=False, remove_short_text=False, lowercase=False, 
                  remove_punctuation=False, remove_stopwords=False, remove_rare_words=False,remove_non_latin=False,
                  rare_limit=5, custom_stopwords=None, min_text_length=3):
    
    # Orijinal veriyi bozmamak i√ßin kopyalƒ±yoruz
    series = series.copy()
    
    if lowercase:
        print("Converting to lowercase...")
        start = timer()
        series = series.str.lower()
        print(f"Completed in {timedelta(seconds=timer() - start)}")

    if remove_hashtag:
        print("Removing hashtags...")
        start = timer()
        series = series.str.replace(r'#\w+', '', regex=True)
        print(f"Completed in {timedelta(seconds=timer() - start)}")

    if remove_mentions:
        print("Removing mentions...")
        start = timer()
        series = series.str.replace(r'@\w+', '', regex=True)
        print(f"Completed in {timedelta(seconds=timer() - start)}")

    if remove_links:
        print("Removing links...")
        start = timer()
        series = series.str.replace(r'http\S+|www\.\S+', '', regex=True)
        print(f"Completed in {timedelta(seconds=timer() - start)}")

    if remove_numbers:
        print("Removing numbers...")
        start = timer()
        series = series.str.replace(r'\d+', '', regex=True)
        print(f"Completed in {timedelta(seconds=timer() - start)}")

    if remove_punctuation:
        print("Removing punctuation...")
        start = timer()
        series = series.str.replace(r'[^\w\s]', '', regex=True)
        print(f"Completed in {timedelta(seconds=timer() - start)}")

    if remove_short_text:
        print(f"Removing short words (length < {min_text_length})...")
        start = timer()
        pattern = r'\b\w{1,' + str(min_text_length - 1) + r'}\b'
        series = series.str.replace(pattern, '', regex=True)
        print(f"Completed in {timedelta(seconds=timer() - start)}")

    # NOT: remove_stopwords=False olduƒüu s√ºrece burasƒ± √ßalƒ±≈ümaz, hata vermez.
    if remove_stopwords:
        print("Removing stopwords...")
        start = timer()
        # stop_kelimeler listesi tanƒ±mlƒ± olmadƒ±ƒüƒ± i√ßin burasƒ± √ßalƒ±≈üƒ±rsa hata verir.
        # Ama duygu analizinde False yaptƒ±ƒüƒ±mƒ±z i√ßin sorun yok.
        stopwords_to_use = custom_stopwords if custom_stopwords else [] 
        series = series.apply(lambda x: ' '.join([word for word in str(x).split() 
                                                   if word.lower() not in stopwords_to_use]))
        print(f"Completed in {timedelta(seconds=timer() - start)}")
    
    if remove_non_latin:
        print("Removing non-latin characters...")
        start = timer()
        series = series.str.replace(r'[^a-zA-Z0-9√ßƒüƒ±√∂≈ü√º√áƒûƒ∞√ñ≈û√ú\s]', '', regex=True)
        print(f"Completed in {timedelta(seconds=timer() - start)}")

    if remove_rare_words:
        print("Removing rare words...")
        start = timer()
        all_words = ' '.join(series.astype(str)).split()
        word_counts = pd.Series(all_words).value_counts()
        rare_words = word_counts[word_counts <= rare_limit]
        rare_words_set = set(rare_words.index)
        series = series.apply(lambda x: ' '.join([word for word in str(x).split() 
                                                   if word not in rare_words_set]))
        print(f"Completed in {timedelta(seconds=timer() - start)}")

    series = series.str.strip().str.replace(r'\s+', ' ', regex=True)
    return series


print("\n--- TEMƒ∞ZLƒ∞K BA≈ûLADI ---")
buyuk_df['Duygu_Icin_Veri'] = preprocessing(
    buyuk_df['Yorum'],
    lowercase=True,          
    remove_links=True,       
    remove_mentions=True,    
    remove_hashtag=True,    
    remove_numbers=True,    
    remove_punctuation=True, 
    remove_stopwords=False   # Kelimeler (ama, fakat, deƒüil) kalsƒ±n √ß√ºnk√º duygu analizi i√ßin √∂nemli olabilir.Ama fakat lakin gibi stop_wordsler
)

```

```python
import torch
from transformers import pipeline
from tqdm import tqdm

#gpu kullanabilmek i√ßin s√ºreyi kƒ±saltmak adƒ±na yazdƒ±m
device = 0 if torch.cuda.is_available() else -1
device_name = torch.cuda.get_device_name(0) if device == 0 else "CPU"

print(f"Model ≈üu cihaz √ºzerinde √ßalƒ±≈üacak: {device_name} (Device ID: {device})")

# 2. MODELƒ∞ Y√úKLEME (GPU'ya)
print("BERT Modeli y√ºkleniyor...")
sentiment_analysis = pipeline(
    "sentiment-analysis", 
    model="savasy/bert-base-turkish-sentiment-cased", 
    device=device,  
    batch_size=64   
)

texts = buyuk_df['Duygu_Icin_Veri'].fillna("").astype(str).tolist()
texts = [text[:512] for text in texts]

results = []
batch_size = 64 

print(f"Toplam {len(texts)} satƒ±r analiz ediliyor...")


for i in tqdm(range(0, len(texts), batch_size)):
    batch_texts = texts[i:i+batch_size]
    try:
       
        batch_results = sentiment_analysis(batch_texts)
        results.extend(batch_results)
    except Exception as e:
        print(f"Batch hatasƒ± (Index {i}): {e}")
        for _ in range(len(batch_texts)):
            results.append({'label': 'NEUTRAL', 'score': 0.0})
buyuk_df['label'] = [res['label'] for res in results]
buyuk_df['score'] = [res['score'] for res in results]

print("\nƒ∞≈ülem Tamamlandƒ±!")
print(buyuk_df[['Duygu_Icin_Veri', 'label', 'score']].head(10))
```
